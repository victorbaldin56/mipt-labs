```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
pressure_data = pd.read_csv("delta_p.csv")
time_data = pd.read_csv("t.csv")
```

```python
delta = pressure_data['d (mum)']
delta_p = pressure_data['dP (kPa)']

plt.figure(figsize=(8, 4))
plt.errorbar(delta_p, delta, xerr=0.5, yerr=0.4, fmt='k.')
[a, b] = np.polyfit(delta_p, delta, deg=1)
x = np.linspace(-10, 10, 1000)
plt.plot(x, a * x + b)

plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle='--')

plt.xlabel(r'$\Delta P$, кПа')
plt.ylabel(r'$\Delta$, мкм')
plt.savefig('pressure.pdf')
```

```python
time_min = time_data['t (min)']
delta_n = time_data['dn (10⁻⁵)']

plt.figure(figsize=(8, 4))
plt.errorbar(time_min, delta_n, xerr=0, yerr=0.4, fmt='k.')
[a, b] = np.polyfit(time_min, delta_n, deg=1)
x = np.linspace(0, 16, 1000)
plt.plot(x, a * x + b)

plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle='--')

plt.xlabel(r'$t$, мин')
plt.ylabel(r'$\delta n$, $10^{-5}$')
plt.savefig('time.pdf')
```
